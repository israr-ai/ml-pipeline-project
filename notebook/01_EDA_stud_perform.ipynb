#### Adding columns for "Total Score" and "Average"

```python
df.head()
```

```python
df['total_score'] = df['math_score'] + df['science_score'] + df['english_score']
df['average'] = df['total_score']/3
df.head()
```

```python
math_full = df[df['math_score'] == 100]['average'].count()
science_full = df[df['science_score'] == 100]['average'].count()
english_full = df[df['english_score'] == 100]['average'].count()

print(f'Number of student with full marks in Maths:{math_full}')
print(f'Number of Students with full marks in science:{science_full}')
print(f'Number of students with full marks in English:{english_full}')
```

```python
science_less_20 = df[df['science_score'] <= 20]['average'].count()
english_less_20 = df[df['english_score'] <= 20]['average'].count()
math_less_20 = df[df['math_score'] <= 20]['average'].count()

print(f'Number of student with less than 20 marks in Maths: {math_less_20}')
print(f'Number of student with less than 20 marks in English : {english_less_20}')
print(f'Number of student with less than 20 marks in Science : {science_less_20}')
```

#### Inights
 
  . From above values we get students have performed the worst in Maths   
  . Best performance is in rading section.
